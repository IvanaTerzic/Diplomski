```python
import folium
import pandas as pd

def plot_data_on_map(report_new, data, final_kol, assessment_kol, log_kol):
    m = folium.Map(location=[data['LAT'].mean(), data['LON'].mean()], zoom_start=10)
    
    # Compute TOKS
    TOKS = []
    for i in range(min(report_new.shape[0], data.shape[0])):  # Loop over valid indices
        assessment_modeli = report_new.loc[[i], assessment_kol].values
        log_modeli = report_new.loc[[i], log_kol].values
        rezultat = assessment_modeli * log_modeli
        rezultat = rezultat.sum()
        TOKS.append(rezultat)
        
        TOC_array = data['TOC(%)'].values
        konc_array = data.iloc[[i]][final_kol].values  # Using iloc for integer-location based indexing

    data['TOKS'] = TOKS
    
    # Plotting
    for i in range(0, len(data)):
        lat = data.iloc[i]['LAT']
        lon = data.iloc[i]['LON']
        TOKS_value = round(data.iloc[i]['TOKS'], 3)
        naziv_postaje = data.iloc[i]['LOKACIJA']
        rijeka = data.iloc[i]['RIJEKA']

        # Generate popup text
        popup_text = f"<b>Rijeka:</b> {rijeka}<br><b>Naziv postaje:</b> {naziv_postaje}<br><b>TOKS:</b> {TOKS_value}"
        
        folium.Circle(
            location=[lat, lon],
            popup=popup_text,
            radius=float(TOKS_value),
            color='blue',
            fill=True,
            fill_color='blue'
        ).add_to(m)
    
    return m

# Sample usage
m = plot_data_on_map(report_new, data, final_kol, assessment_kol, log_kol)
m
# m.save("map.html")
```
